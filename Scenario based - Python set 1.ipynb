```python
# Create a list or given list of numbers
# To find the largest number in the list ---> define it
# Take initial values in the list is the largest
# and go through each number in the list to check if it's greater than our current largest number
# once find it and hold the largest number in the list and return
# finally execute it.
```

```python
numbers = [12, 45, 7, 23, 56, 89, 34]

def find_largest_manual(numbers):
 
    largest = numbers[0]
    for num in numbers:
        if num > largest:
            largest = num
    return largest
```

```python
# 3.	Scenario: A company provides employees with a 10% bonus if their salary exceeds $50,000.
 Write logic to determine the bonus amount based on the given salary.

# Read the Employee salary     
# To determine the bonus amount
# If salary exceeds the 50000,10% bonus is added
# If not no bonus to be added.
# Calculated the bonus amount and Return it.
# Calculate the total amount after the bonus amount to be added (If you need)  
```

```python
def calculate_bonus(salary):
    """Calculates 10% bonus if salary exceeds $50,000"""
    if salary > 50000:
        bonus = 0.10 * salary
        return bonus
    else:
        return 0

# Example usage
salary = float(input("Enter the salary in $: "))
bonus = calculate_bonus(salary)
print(f"Bonus amount: ${bonus:.2f}")
print("Total Amount in $:",bonus+salary) # I've included
```

```python
# 5.Scenario: A text-processing tool reverses a given word or sentence for formatting purposes.

# 
```

```python
word="Reverse the Sentence"[::-1]
print(word)

# Assigned the sentence to the variable name word
# Reverse the Sentence"[::-1] by using Slicing
# And print the word

# code takes the string "Reverse the Sentence", reverses it character by character, and stores the reversed result in the variable word
```

```python
def reverse_text(text):
    return text[::-1]
def main():
    print("Welcome to the Text Reversal Program!")
    print("Enter 'exit' at any time to quit.")

    while True:
        user_input = input("Enter a word or sentence: ")

        if user_input.lower() == 'exit':
            print("Exiting program. Goodbye!")
            break

        reversed_string = reverse_text(user_input)
        print(f"Original: '{user_input}'")
        print(f"Reversed: '{reversed_string}'")
        print("-" * 30)

if __name__ == "__main__":
    main()
```

```python
# A retail store offers a 20% discount if a customer’s total order exceeds $100. Write logic to calculate the final amount to be paid after applying the discount

# Read the total putchased amount
# If the amount exceeds $100 calculate the discount is 20%
# Subtract the discount price from the original price to get the amount to be paid after discount
```

```python
# Input values

price = float(input("Enter the original price: "))
discount = float(input("Enter discount percentage: "))
discount_amount = (price * discount) / 100  # Calculate discount amount
final_price = price - discount_amount   # Calculate final price after discount

print(f"Original Price: ${price:.2f}")
print(f"Discount Percentage: {discount}%")
print(f"Discount Amount: ${discount_amount:.2f}")
print(f"Final Price after Discount: ${final_price:.2f}")
```

```python
year = int(input("Enter the Year:"))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is a common year.")
```

```python
# 10.	Scenario: A program filters out only even numbers from a given list.
 Write logic to extract and return only the even numbers from a list

# Read the numbers frpm the list
# Create the empty list to store even number
# Go through the list and check each number is divisible by 2
# Once found all the even numbers and return to empty list to store
# print the output and even number is extracted from the list
```

```python
numbers = [20, 10, 16, 19, 25, 1, 276, 188]
for num in numbers:
    if num % 2 == 0:
        print(f"{num} is even")
```

```python
def get_even_numbers(numbers_list):
    even_numbers = []
    for num in numbers_list:
        if num % 2 == 0:
            print(f"{num} is even")
            even_numbers.append(num)
    return even_numbers # The return statement goes here!

numbers = [20, 10, 16, 19, 25, 1, 276, 188]
result = get_even_numbers(numbers)
print(result) # This would print [20, 10, 16, 276, 188]
```
