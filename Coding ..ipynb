```python
5.Write a Python program to check if a number is prime. 
    Expected Output (for input 7):
7 is a prime number

---Doubt---
```

```python
def is_prime(n):
    factors = 0
    for i in range(1, n+1):
        if n % i == 0:
            factors += 1
    return factors == 2

num = 7
if is_prime(num):
    print(f"{num} is a prime number")
else:
    print(f"{num} is not a prime number")
```
